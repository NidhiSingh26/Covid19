Importing Covid library


```python
from covid import Covid
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
covid = Covid()
covid.get_data()
```

```python
name = input("Enter the country Name:-")
```

```python
virusdata = covid.get_status_by_country_name(name)
```

```python
remove = ['id','country','latitude','longitude','last_update']
```

```python
for i in remove:
  virusdata.pop(i)
```

```python
all = virusdata.pop('confirmed')
id = list(virusdata.keys())
value = [str(i) for i in virusdata.values()]
```

Get Status By Country Name

```python
covid.get_status_by_country_name("india")
```

```python
plt.pie(value, labels= id, colors = ['r','y','g','b'] ,autopct= '%1.1f%%')
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='active')  
plt.plot(x, x**2, label='recovered')  
plt.plot(x, x**3, label='death')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("COUNTRY : "+name.upper() +"\n TOTAL CASES : "+str(all))
plt.legend()
plt.show()
```
